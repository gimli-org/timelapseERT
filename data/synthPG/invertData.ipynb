```python
import matplotlib.pyplot as plt
import pygimli as pg
from pygimli.physics import ert
```

```python
tl = ert.TimelapseERT("synth.shm")
print(tl)
```

```python
tl.fullInversion(reg=dict(correlationLengths=[10, 1]))
```

```python
_, ax = plt.subplots(nrows=5, ncols=2, sharex=True, sharey=True, figsize=(8, 8))
for i, model in enumerate(tl.models):
    pg.show(tl.pd, model, cMin=10, cMax=400, cMap="Spectral_r", 
            logScale=True, colorBar=False, ax=ax.flat[i])
```
