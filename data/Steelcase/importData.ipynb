```python
import pygimli as pg
from pygimli.physics import ert
```

```python
from glob import glob
from datetime import datetime
filenames = glob("data/*.dat")
```

```python
times = [datetime.fromisoformat(fn[15:-4]) for fn in filenames]
```

```python
mue = ert.TimelapseERT("data/*.dat", times=times)
```

```python
mue.showData(t=1, cMin=20, cMax=100, crossplot=True)
```

```python
mue.generateDataPDF(cMin=20, cMax=100, crossplot=True)
```
